### 1
Ответьте на вопросы:
- В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?
- Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы

1. Схожесть заключается в получении результата предсказания в виде произведения эмбеддинга юзера и эмбеддинга айтема. Различие заключается в способе получения эмбеддингов. При CF эмбеддинги обучаются при матричной факторизации, а в гибридных системах (lightFM) в результате суммирования эмбеддингов признаков, которыми описываются юзеры и айтемы.
2. Гибридные системы необходимы в задачах, где есть проблемы холодного старта или сильно разреженной матрицы интеракций. 
Примерами таких задач являются: а) Задача, описанная в статье lightFM, магазину одежды необходимо рекомендовать новые айтемы, которые могут быть хорошо описаны фичами, но для которых нет достаточной истории взаимодействия с пользователями, так как товар новый. б) Задача, которую я решал на хакатоне. Необходимо рекомендовать закупки для пользователя. При этом для новых закупок не может быть матрицы интеракций с пользователями. так ка кэто закрытая информация даже для самой площадки. Таким образом рекомендации могут даваться только исходя из близости закупок, определяемой по ее фичам.
