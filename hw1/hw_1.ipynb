1) Сравните метрики hit_rate@k, precision@k. Какую матрику использовать предпочтительно и почему. Привидите пример 2-3 задач (опишите, что является клиентом, что товаром), в которой более уместно использовать метрику hit_rate?

Hit rate принимает значение 1, если хотя бы 1 товар из рекомендованных оказался релевантным. precision@k учитывает количество релевантных товаров в составе рекомендованных. Хит рейт можно использовать в задачах , где предложение списка с 1 релевантным товаром является достаточным. 
Допустим видео хостинг предлагает 5 видео для просмотра юзеру, и так как одновременно все 5 видео смотреть бессмысленно, то ситуация, в которой 1 видео заецпит юзера и он перейдет и посмотрит его - хороший результат.
ТОже можно сказать про онлайн-кинотеатры, если из К фильмов человеку понравится хотя бы 1, то он его купит, что можно считать успешным кейсом. 
(Клиент - человек, товар - видео и фильмы)
